```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.3
)
```

```python
#Detect the type of emergency
@tool
def emergency_detector(query: str) -> str:
    """Detects the type of emergency from user message."""

    return """
🚨 Emergency Type Detected:
Possible medical emergency.
Immediate attention required.
"""
```

```python
#Provide first aid instructions
@tool
def first_aid_guide(query: str) -> str:
    """Provides first aid instructions for emergency situations."""

    return """
🩺 First Aid Instructions:
1. Check breathing and pulse.
2. Keep the person lying down.
3. Loosen tight clothing.
4. If unconscious but breathing, place in recovery position.
5. Call emergency services immediately.
"""
```

```python
#Suggest nearby hospitals
@tool
def hospital_locator(query: str) -> str:
    """Provides nearby hospital information during an emergency."""

    return """
🏥 Nearby Hospitals:
1. City Care Hospital – 2.5 km
2. Apollo Emergency Center – 4 km
3. Government General Hospital – 3 km
"""
```

```python
#Generate an SOS emergency message
@tool
def sos_generator(query: str) -> str:
    """Generates an emergency SOS alert message for immediate help."""

    return """
📢 SOS MESSAGE:

This is an emergency. A person requires immediate medical assistance.
Location: [User Location]
Please send help urgently.
"""
```

```python
llm_with_tools = llm.bind_tools([
    emergency_detector,
    first_aid_guide,
    hospital_locator,
    sos_generator
])
```

```python
# Get query from user
query = input("Enter your emergency situation: ")

# Send query to LLM
response = llm.invoke(query)

# Print response
print(response.content)
```
